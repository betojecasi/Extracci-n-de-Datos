**Filtros basicos para variables cuantitativas**

```python
#Filtro por comparacion
filtro6=df[df["independence_year"]> 1900]
filtro6
#Aqui filtramos los que de independizaron despues de 1900, para saber que tan recientemente se independizaron
#ciertos paises
```

```python
#Filtro Y
filtro7=df[(df['independence_year']>1900) & (df['continent']=='Europe')]
filtro7
#Aqui hacemos lo mismo solo que especificamente de los paises de Europa y los paises con independecia despues de 1900
```

```python
filtro7.to_csv("Filtro7.csv")
```

```python
from google.colab import files

files.download("Filtro7.csv")
```

```python
#Filtro o
filtro8=df[(df["independence_year"]>1900) | (df["continent"]=="Africa")]
filtro8
#Aqui se incluyen los paises con independencia despues de 1900 o los del continente Africa
```

```python
filtro8.to_csv("Filtro8.csv")
```

```python
#Filtro not
filtro9=df[~(df["independence_year"]>1900) & ~ (df["continent"]=="North America")]
filtro9
#Aqui no filtra los datos que se ingresan sino que los evade (nu muestra a norte america ni a los paises con independecia >1900)
```
